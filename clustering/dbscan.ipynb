```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
% matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
irisData=iris.data
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.6,min_samples=4)
dbscan.fit(irisData)
labels=dbscan.labels_
```

```python
# draw scatter plot
plt.scatter(irisData[:,2],irisData[:,3],c=labels)
plt.show()
```
