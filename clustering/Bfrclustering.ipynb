```python
from scipy.cluster.hierarchy import dendrogram,linkage,fcluster
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris_data=iris.data
```

```python
link=linkage(iris_data,method='complete')
dnd=dendrogram(link)
```

```python
labels=fcluster(link,4,criterion='distance')
```

```python
# show it
plt.scatter(iris_data[:,2],iris_data[:,3],c=labels)
plt.show()
```
