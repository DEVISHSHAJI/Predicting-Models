```python
orders=pd.read_csv('/content/orders.csv')
```

```python
orders.head()
```

```python
orders.info()
```

```python
transaction_digi=orders.groupby('ID_Customer')
trans_items=[]
for transaction_digi,i in transaction_digi:
      trans_items.append(i['ID_Item'].values.tolist())
trans_items
```

```python
trans_items_new = []
for tr in trans_items:
    if len(tr) > 1:
        trans_items_new.append(tr)


pattern = pyfpgrowth.find_frequent_patterns(trans_items_new,3)
{k: pattern[k] for k in list(pattern)[:50]}
```

```python
rules = pyfpgrowth.generate_association_rules(pattern, 0.25)
rules
```
